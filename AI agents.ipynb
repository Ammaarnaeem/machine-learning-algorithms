```python
#INSTRUCTIONS
#Enter LOCATION A/B in captial letters
#Enter Status O/1 accordingly where 0 means CLEAN and 1 means DIRTY

def vacuum_world():
    # Initialize goal state with both rooms as clean
    
#     goal_state = {'A': '0', 'B': '0'}
    cost = 0

    # Get user input for vacuum location and room status
    location = input("Enter location of vacuum (A/B): ").upper()
    status = input(f"Enter status of room {location} (0/1): ")
    other_room = 'A' if location == 'B' else 'B'
    other_status = input(f"Enter status of room {other_room} (0/1): ")

    goal_state = {location: status, other_room: other_status}
 
    print("location condition:", goal_state)

    if status == '1':
        # Vacuum is in a dirty room
        goal_state[location] = '0'
        cost += 1
        print(f"Location {location} cleaned. Cost: {cost}")
        
        
        if other_status == '1':
            # Other room is dirty
            print(f"Moving to room {other_room}")
            cost += 1
            goal_state[other_room] = '0'
            cost += 1
            print(f"Location {other_room} cleaned. Cost: {cost}")
        else:
            print(f"Room {other_room} is already clean")
    else:
        print(f"Location {location} is already clean")
        
        if other_status == '1':
            # Other room is dirty
            print(f"Moving to room {other_room}")
            cost += 1
            goal_state[other_room] = '0'
            cost += 1
            print(f"Location {other_room} cleaned. Cost: {cost}")
        else:
            print(f"Room {other_room} is already clean")

    print("\nGoal state:", goal_state)
    print("Performance measurement:", cost)

vacuum_world()
```

```python
import logging

class SoilMoistureSensor:
    def __init__(self, dry_threshold, moist_threshold, wet_threshold):
        self.dry_threshold = dry_threshold
        self.moist_threshold = moist_threshold
        self.wet_threshold = wet_threshold

    def measure_moisture(self):
        try:
            moisture_reading = float(input("Enter the moisture level: "))

            if moisture_reading <= self.dry_threshold:
                return "Dry"
            elif moisture_reading <= self.moist_threshold:
                return "Moist"
            elif moisture_reading <= self.wet_threshold:
                return "Wet"
            else:
                raise ValueError("Invalid moisture level")
        except Exception as e:
            logging.error("Error measuring soil moisture: %s", e)
            raise

class PlantWateringControl:
    def __init__(self, moisture_sensor):
        self.moisture_sensor = moisture_sensor

    def execute(self):
        soil_moisture_state = self.moisture_sensor.measure_moisture()

        if soil_moisture_state == "Dry":
            self._water_plants()
        elif soil_moisture_state == "Moist":
            self._do_nothing()
        elif soil_moisture_state == "Wet":
            self._stop_watering()
        else:
            raise ValueError("Invalid soil moisture state")

    def _water_plants(self):
        print("Watering plants...")

    def _do_nothing(self):
        print("Soil moisture level is adequate, doing nothing...")

    def _stop_watering(self):
        print("Soil moisture level is high, stopping watering...")

if __name__ == "__main__":
    logging.basicConfig(filename="plant_watering.log", level=logging.ERROR)

    dry_threshold = 20.0
    moist_threshold = 40.0
    wet_threshold = 60.0

    moisture_sensor = SoilMoistureSensor(dry_threshold, moist_threshold, wet_threshold)
    watering_control = PlantWateringControl(moisture_sensor)
    watering_control.execute()
```
